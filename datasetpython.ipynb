```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data = pd.read_csv('C:\\Users\\sec\\Documents\\Sprint 1 project\\data%20set.csv')  # Adjust the path and file name as necessary

# Create a bar plot
plt.figure(figsize=(10, 6))
bar_plot = sns.barplot(x='Device Model', y='User Behavior Class', data=data, palette='viridis')

# Adding titles and labels
plt.title("Analyzing Mobile User Behavior Patterns")
plt.xlabel("Device Model")
plt.ylabel("User Behavior Class")
plt.xticks(rotation=45)

# Adding data labels on top of each bar
for p in bar_plot.patches:
    bar_plot.annotate(f'{p.get_height():.2f}',  # Format the height to two decimal places
                      (p.get_x() + p.get_width() / 2., p.get_height()),  # Position at center top of bar
                      ha='center', va='bottom',  # Align text above the bar
                      fontsize=10, color='black')  # Text styling

# Adjust layout to make room for the labels
plt.tight_layout()
plt.show()
```

```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = ('C:\\Users\\sec\\Documents\\Sprint 1 project\\data%20set.csv')  # Update with the path to your file
data = pd.read_csv(file_path)

# Display basic info to understand the dataset structure
print(data.info())
print(data.head())

# Example 1: Bar Chart - Building predictive models for app usage
top_apps = data.groupby('Device Model')['App Usage Time (min/day)'].sum().nlargest(10)
plt.figure(figsize=(12, 6))
ax = top_apps.plot(kind='barh', color='skyblue')
plt.title("Building predictive models for app usage")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Device Model")

# Add data labels
for index, value in enumerate(top_apps):
    plt.text(value, index, f'{value:.2f}', va='center')  # Label positioned horizontally

plt.show()

# Example 2: Line Chart - Data Usage over Time
plt.figure(figsize=(12, 6))
ax = sns.lineplot(x='Device Model', y='App Usage Time (min/day)', data=data, color='purple')
plt.title("Developing insights into battery and data consumption trends")
plt.xlabel("Device Model")
plt.ylabel("App Usage Time (min/day)")


plt.xticks(rotation=45)
plt.show()

# Example 3: Histogram - Gender Wise Custom Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Gender'], bins=30, kde=True, color='green')
plt.title("Gender Wise Custom Distribution")
plt.xlabel("Gender")
plt.ylabel("Frequency")

# Histograms typically donâ€™t include individual data labels on bars, but for frequency, 
# you can add counts if required using this alternative approach:
# Adding labels could lead to cluttering due to bin overlap

plt.show()

# Example 4: Scatter Plot - Data Usage vs. Duration
plt.figure(figsize=(10, 6))
ax = sns.scatterplot(x='App Usage Time (min/day)', y='Data Usage (MB/day)', data=data, hue='App Usage Time (min/day)', palette='viridis')
plt.title("App Usage vs. Data Usage")
plt.xlabel("App Usage Time (min/day)")
plt.ylabel("Data Usage (MB/day)")


plt.legend(loc='upper right')
plt.show()
```
